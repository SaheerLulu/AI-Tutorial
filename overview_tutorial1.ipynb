```python
def update_parameter(a,b):
    der_loss_a = (loss(a=a+h,b=b) - loss(a=a,b=b))/h
    der_loss_b = (loss(a=a,b=b+h) - loss(a=a,b=b))/h
    a = a-0.025*der_loss_a
    b = b-0.025*der_loss_b
    return a,b
    
```
